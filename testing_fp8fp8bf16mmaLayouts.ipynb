```python
from layout_module import Layout, NestedTuple 
from layout_module import layout_algebra as la
from layout_module import flat_algebra as fa
from layout_module import nested_tuple_algebra as na
```

```python
s1 = NestedTuple.from_literal((32,2,2))
d1 = NestedTuple.from_literal((2,64,1))

wC_frag = Layout(s1,d1)
s2 = NestedTuple.from_literal((16,8))
d2 = NestedTuple.from_literal((8,1))
logical_C_frag = Layout(s2,d2)
```

```python
wC = la.compose(wC_frag, logical_C_frag)
```

```python
coordinate_map = fa.coordinate_map
colex_inv_map = fa.colex_inv
```

```python
map = [None]*wC.cosize
for x in range(wC.size): 
  off = fa.layout_map(wC.shape.int_tuple, wC.stride.int_tuple, x)
  coord = fa.colex_inv(wC.shape.int_tuple,x)
  coord_sub = na.substitute(wC.shape.prof, [NestedTuple.from_literal(coord[i]) for i in range(len(coord))])
  map[off] = coord_sub
```
