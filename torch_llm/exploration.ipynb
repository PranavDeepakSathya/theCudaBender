```python
import torch
device = "cuda"
import time 
```

```python
A,B = torch.randn(4096,4096, dtype = torch.bfloat16, device=device), torch.randn(4096,4096, dtype = torch.bfloat16, device=device)
C = torch.zeros(4096,4096, dtype = torch.float32, device = device)
```

```python
flop = 2*4096*4096*4096
```

```python
# warmup
for _ in range(5):
    C = torch.matmul(A, B)
torch.cuda.synchronize()

s = time.perf_counter()

for _ in range(10000):
    C = torch.matmul(A, B)

torch.cuda.synchronize()
e = time.perf_counter()
```

```python
time_measured = ((e-s)/10000) 
tflops = (flop/time_measured)/(1e12)
```
