```python
import torch
```

```python
words = open('names.txt','r').read().splitlines()
```

```python
words[:10]
```

```python
chars = sorted(list((set(''.join(words)))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {y:x for x,y in stoi.items()}
```

```python

smoothing_factor = 2 
bigram = torch.zeros(27,27) + smoothing_factor

for w in words: 
  chs = ['.'] + list(w) + ['.']
  for ch1,ch2 in zip (chs,chs[1:]):
    bigram[stoi[ch1],stoi[ch2]] +=1
    
    
```

```python
probs = (bigram)/(torch.sum(bigram, dim = 1, keepdim=True))
print(probs[0].sum())
```

```python
def sample_name(probs,stoi,itos, max_length = 500): 
  name = []
  c = '.'
  for _ in range(max_length):
    prob_row = probs[stoi[c]]
    sampled_idx = torch.multinomial(prob_row, 1, replacement=True)
    new_c = itos[sampled_idx.item()]
    if new_c == '.':
      break
    name.append(new_c)
    c = new_c
    
  return ''.join(name)
```

```python
names = [sample_name(probs, stoi, itos) for _ in range(10)]
```

```python
#calculate log likelihood and add it up over the whole dataset 
#to maximize likelihood, maximize log likelhood, or minimize negative log likelihhod 
#well minimize its avg
loss = 0
for w in words: 
  chs = ['.'] + list(w) + ['.']
  for ch1,ch2 in zip (chs,chs[1:]):
    loss += -torch.log(probs[stoi[ch1],stoi[ch2]])
```

```python
avg_loss = loss/len(words)
```
