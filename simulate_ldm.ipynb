```python
import numpy as np
```

```python
def simulate_compact_ldmatrix_x4(A, lane_bjiection):
  assert np.size(A) == 16*32 
  assert np.size(np.unique(lane_bijection)) == (np.size(lane_bijection))
  for l in lane_bijection: 
    assert 0 <= l 
    assert l < 32 
    
  expanded_bij = lane_bijection*16 
  
  return_shape = (4,32,4) #L, N_reg, N_pack_per_reg 
  ret = np.zeros(return_shape).astype(np.int32)
  for i,lane_addr in enumerate(expanded_bij):     
    
    for j in range(16): 
      linear_idx = i*16  + j 
      linear_addr = lane_addr + j 
      pack_id = linear_idx % 4
      lane_id = (linear_idx //4) % 32 
      reg_id = i//8
      ret[reg_id,lane_id,pack_id] = A[linear_addr]
      
  return ret
      
    
  
```

```python
A = np.arange(16*32).astype(np.int32)
lane_bijection = (3*(np.arange(32)) + 7)%32
```

```python
V = simulate_compact_ldmatrix_x4(A,lane_bijection)
```
